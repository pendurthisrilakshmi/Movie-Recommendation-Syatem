Importing libraries

```python
import numpy as np
import pandas as pd
import difflib
# if movie name is misspelt it will get close match
from sklearn.feature_extraction.text import TfidfVectorizer
#used to convert textual data to numerical data(feature vectors)
from sklearn.metrics.pairwise import cosine_similarity
#gives similarity score of all movies compared to others
```

Data collection and pre-processing

```python
#loading data
movies_data=pd.read_csv('movies (1).csv')
```

```python
# number of rows and columns in the data set

movies_data.shape
```

```python
# printing the first 10 rows
movies_data.head()
```

```python
# feature selection

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replacing the null/missing values with null string in the selected features

for i in selected_features:
  movies_data[i] = movies_data[i].fillna('')
  #fillna() replaces null values with specified value
```

```python
# combining all the selected features

combining_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
print(combining_features)#concatenating selected features for all movies
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
#bcoz cosine similarity is easy to perform on numerical values ratherthan text
```

```python
feature_vectors = vectorizer.fit_transform(combining_features)
# feature_vector store all numerical values
print(feature_vectors)
```

Cosine Similarity

```python
# calculating the similarity scores using cosine similarity
similarity_scores = cosine_similarity(feature_vectors)
print(similarity_scores)
#prints similarity scores of all movies with respect to each movie
print(similarity_scores.shape)
```

Getting the movie name from the user

```python
# creating a list with all the movie names given in the dataset
list_of_all_titles = movies_data['original_title'].tolist()
print(list_of_all_titles)
```

```python
#getting name of the movie from user
movie_name = input(' Enter the name of the movie you wish to watch : ')
```

```python
# finding the close match for the movie name given by the user

finding_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(finding_close_match)
```

```python
close_match = finding_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity_scores[index_of_the_movie]))
#enumerate to run a for loop in particular list
print(similarity_score)
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
#sorting in descending order.So reverse=true.
#x[1] is similarity scores values
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for j in sorted_similar_movies:
  index = j[0]
  title_using_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<21):
    print(i, '.',title_using_index)
    i+=1
```
